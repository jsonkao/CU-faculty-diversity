# Faculty Diversity in Sex

```python
%%capture

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import json

matplotlib.rcParams.update({'font.size': 14})
plt.tight_layout()
```

```python
xls = pd.ExcelFile('../data/gender.xlsx')

dfs = dict(pd.read_excel(xls, sheet_name=None))
```

```python
allDivisions = [ 'HUM', 'NAT', 'SOC', 'ENG' ]
divisions = {
    'HUM': 'Humanities',
    'NAT': 'Natural Sciences',
    'SOC': 'Social Sciences',
    'ENG': 'Engineering',
}

allStatuses = [ 'NE', 'EL', 'TE' ]
statuses = {
    'NE': 'not eligible for tenure',
    'EL': 'tenure-track',
    'TE': 'tenured',
}
```

```python
def lineGraph(ax, div, status, abbrev=False, showPercents=False):
    if isinstance(div, int) and isinstance(status, int):
        div = allDivisions[div]
        status = allStatuses[status]
        
    df = dfs['{}-{}'.format(div, status)]
    name = '{}, {}'.format(
        divisions[div],
        status if abbrev else statuses[status]
    )
    
    if showPercents:
        ax.plot(df['Year'], df['% Women'])
        ax.axis(ymin=0, ymax=1)
    else:
        ax.plot(df['Year'], df['Women'])
        ax.plot(df['Year'], df['Total'])
        ax.axis(ymin=0, ymax=175)
    ax.grid()
    ax.legend()
    ax.set(xlabel='Year', ylabel='Women', title=name)
```

```python
def graphAllPercents():
    fig, ax = plt.subplots(4, 3, figsize=(21, 28))

    for i in range(12):
        lineGraph(ax[i//3, i%3], i // 3, i % 3, showPercents=True)
```

Let's observe female representation in every division through the tenure pipeline.

```python
graphAllPercents()
```

```python
def graphAllRaw():
    fig, ax = plt.subplots(4, 3, figsize=(21, 28))

    for i in range(12):
        lineGraph(ax[i//3, i%3], i // 3, i % 3)
```

```python
graphAllRaw()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(21, 14))

for i in range(6):
    lineGraph(ax[i//3, i%3], 1, i % 3, showPercents=(i>2))
```

The three graphs for each academic division look like stairs stepping down. The NE and TE of each division would be good for you-draw-it's. Let's export this data.

```python
M = {}
for i in range(12):
    div = allDivisions[i // 3]
    status = allStatuses[i % 3]
    name = '{}-{}'.format(div, status)
    df = dfs[name]
    M[name] = list(df.T.to_dict().values())
print(json.dumps(M))
```
