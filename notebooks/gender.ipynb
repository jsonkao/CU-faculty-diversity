# Faculty Diversity in Sex

```python
%%capture

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update({'font.size': 14})
plt.tight_layout()
```

```python
xls = pd.ExcelFile('../data/gender.xlsx')

dfs = dict(pd.read_excel(xls, sheet_name=None))
```

```python
divisions = [ 'Humanities', 'Natural Sciences', 'Social Sciences' ]
statuses = [ 'Non-tenured', 'Tenure-track', 'Tenured']
```

```python
def lineGraph(ax, div, status):
    name = '{}, {}'.format(divisions[div], statuses[status])
    df = dfs[name]
    ax.plot(df['Year'], df['% Women'])
    ax.grid()
    ax.axis(ymin=0, ymax=1)
    ax.set(xlabel='Year', ylabel='% Women', title=name)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(21, 21))

for i in range(9):
    lineGraph(ax[i//3, i%3], i // 3, i % 3)
    
plt.savefig('../img/steps.png')
```
